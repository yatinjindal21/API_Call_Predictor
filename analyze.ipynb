```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
api_data = pd.read_csv('separate_data/A2.csv')
api_data['call_time'] = pd.to_datetime(api_data['call_time'])

api_data['day'] = api_data['call_time'].dt.day

data_day = api_data.groupby(['day']).size().reset_index(name='freq')

plt.plot(data_day['day'],data_day['freq'])
```

```python
api_data['hour'] = api_data['call_time'].dt.hour

data_hour = api_data.groupby(['hour']).size().reset_index(name='freq')

plt.plot(data_hour['hour'],data_hour['freq'])
```

```python
api_data['weekday'] = api_data['call_time'].dt.day_name()

data_hour = api_data.groupby(['weekday']).size().reset_index(name='freq')

plt.plot(data_hour['weekday'],data_hour['freq'])
```

```python
data_hour = api_data.groupby(['api_name']).size().reset_index(name='freq')

plt.plot(data_hour['api_name'],data_hour['freq'])
```

```python
api_data['call_time'] = pd.to_datetime(api_data['call_time'])

# Set 'call_time' as the index
api_data.set_index('call_time', inplace=True)

monthly_data = api_data.resample('M').size().to_frame('call_count')

monthly_data.reset_index()
monthly_data
plt.figure(figsize=(20, 5))
plt.plot(monthly_data.index, monthly_data['call_count'])
```
